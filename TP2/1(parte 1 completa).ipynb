```python
df_patagonia_dta = df[df["region"] == "Patagónica"] # UNA VEZ LOCALIZADA, RAELIZAMOS EL FILTRO
df_patagonia_dta.head() #CORROBORAMOS QUE ESTÉ CORRECTO
```

```python
df_xlsx = pd.read_excel("usu_individual_T124.xlsx") #REPETIMOS TODO LO ANTERIOR PERO EN FORMATO XLSX
df_xlsx.head()
```

```python
df_xlsx["REGION"].unique()
```

```python
#COMO LA BASE .xlsx INDICA LAS REGIONES CON NÚMEROS, LAS RENOMBRAMOS SEGÚN DEL DICCIONARIO PARA MEJORAR LA LECTURA
region_dict = { 
    1: "Gran Buenos Aires",
    40: "Noroeste",
    41: "Noreste",
    42: "Cuyo",
    43: "Pampeana",
    44: "Patagonia"
}
```

```python
df_xlsx["REGION"] = df_xlsx["REGION"].replace(region_dict) 
```

```python
df_xlsx["REGION"].unique()
```

```python
#REALIZAMOS EL FILTRO PERO AHORA APLICADO AL .xlsx
df_patagonia_xlsx = df_xlsx[df_xlsx["REGION"] == "Patagonia"]
df_patagonia_xlsx.head()
```

```python
df_patagonia_dta.loc[:, "region"] = df_patagonia_dta["region"].replace("Patagónica", "Patagonia") #Puesto que "Patagonia" no es lo mismo que "Patagónica", procedemos a unificar todo a "Patagonia"
df_patagonia_dta.head() #Visualizamos
```

```python
df_patagonia_dta.columns = df_patagonia_dta.columns.str.upper() #Unificamos también las mayúsculas en ambos documentos
df_patagonia_dta.head() #Visualizamos
```

```python
#Visualizo todos los encabezados de columasn 
print(list(df_patagonia_dta.columns))   
print(list(df_patagonia_xlsx.columns))  
```

```python
columnas_deseadas = [
    'CODUSU', 'NRO_HOGAR', 'COMPONENTE', 'H15', 'ANO4', 'TRIMESTRE', 'REGION',
    'MAS_500', 'AGLOMERADO', 'PONDERA', 'CH03', 'CH04', 'CH06', 'CH07', 'CH08',
    'CH09', 'CH10', 'CH11', 'CH12', 'CH13', 'CH14', 'CH15', 'NIVEL_ED', 'ESTADO',
    'CAT_INAC', 'IPCF'
]
```

```python
df_patagonia_dta_filtrado = df_patagonia_dta[columnas_deseadas]
df_patagonia_xlsx_filtrado = df_patagonia_xlsx[columnas_deseadas]
```

```python
df_unido = pd.concat([df_patagonia_dta_filtrado, df_patagonia_xlsx_filtrado], ignore_index=True)
```

```python
print(df_unido.shape)        # Vemos la cantidad de filas y columnas
print(df_unido.columns)      # Verificamos la lista y orden de columnas
df_unido.head()              # Vemos las primeras filas para corroborar que este okey 
```

correcto. Selecciones 15 variables de interés y reporten la 
cantidad de valores faltantes (NA, o NaN en Python) en una 
tabla por cada año. Comenten qué variables de las 15 que 
seleccionarion tienen más valores faltantes y qué año.****

```python
import numpy as np
```

```python
#Seleccionamos al azar 15 variables de las 26 que dejamos anteriormente y las escribimos en una lista
variables_interes = [
    'ANO4','CH03', 'CH04', 'CH06', 'CH07', 'CH08', 'CH09', 'CH10', 'CH11',
    'CH12', 'CH13', 'CH14', 'CH15', 'NIVEL_ED', 'ESTADO', 'IPCF'
]
```

```python
df_interes = df_unido[variables_interes]
```

```python
#Definimos  los valores que consideraremos como errores (exceptuando las columnas que aceptan 0)
valores_erroneos = {
    'CH03': [0, '', '9999'],  
    'CH04': [0, '', '9999'],  
    'CH06': [0, '', '9999'], 
    'CH07': [0, '', '9999'], 
    'CH08': [0, '', '9999'],  
    'CH09': [0, '', '9999'],  
    'CH10': [0, '', '9999'],  
    'CH11': [0, '', '9999'], 
    'CH12': [0, '', '9999'],  
    'CH13': [0, '', '9999'], 
    'NIVEL_ED': ['', '9999'],
    }

for columna, valores in valores_erroneos.items():
    df_interes.loc[:, columna] = df_interes[columna].replace(valores, np.nan)
```

```python
na_por_ano = df_interes.groupby('ANO4').agg(lambda x: x.isna().sum())
print(na_por_ano)
na_por_ano
```

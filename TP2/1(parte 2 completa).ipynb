****3. Realicen un gráfico de barras mostrando la composición por sexo para 
2004 y 2024. Comenten los resultados.****

```python
import matplotlib.pyplot as plt
```

```python
df_2004 = df_unido[df_unido['ANO4'] == 2004]
df_2024 = df_unido[df_unido['ANO4'] == 2024]
```

```python
def mapear_sexo(valor):
    if valor in [1, '1', 'Varón']:
        return 'Varón'
    elif valor in [2, '2', 'Mujer']:
        return 'Mujer'
    else:
        return 'Otro/Desconocido'
```

```python
conteo_2004 = df_2004['CH04'].map(mapear_sexo).value_counts()
conteo_2024 = df_2024['CH04'].map(mapear_sexo).value_counts()
```

```python
df_conteo = pd.DataFrame({
    '2004': conteo_2004,
    '2024': conteo_2024
}).fillna(0)
```

```python
ax = df_conteo.plot(kind='bar', figsize=(8, 5), color=['skyblue', 'lightgreen'])
plt.title('Composición por sexo en 2004 y 2024')
plt.xlabel('Sexo')
plt.ylabel('Cantidad de personas')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar el número exacto de cada sexo
for p in ax.patches:
    altura = p.get_height()
    if altura > 0:
        ax.annotate(f'{int(altura)}', (p.get_x() + p.get_width() / 2, altura + 50),
                    ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_104 = pd.read_stata("C:/Users/Marco/Downloads/usu_individual_T104.dta")
df_124 = pd.read_excel("C:/Users/Marco/Downloads/usu_individual_T124.xlsx")


df_104["AÑO"] = 2004
df_124["AÑO"] = 2024

df = pd.concat([df_104, df_124], ignore_index=True)

respondieron = df[df['ESTADO'] != 0].copy()
norespondieron = df[df['ESTADO'] == 0].copy()

print(f"Cantidad de personas que NO respondieron la condición de actividad: {norespondieron.shape[0]}")
print(f"Cantidad de personas que SÍ respondieron: {respondieron.shape[0]}")

# Crear variable PEA (1 si ESTADO es 1 u 2)
respondieron['PEA'] = respondieron['ESTADO'].apply(lambda x: 1 if x in [1, 2] else 0)

# Gráfico de barras PEA por año
pea_por_anio = respondieron.groupby(['AÑO', 'PEA']).size().unstack().fillna(0)
pea_por_anio.plot(kind='bar', stacked=True)
plt.title("Composición de la Población Económicamente Activa (PEA) por año")
plt.xlabel("Año")
plt.ylabel("Cantidad de personas")
plt.legend(['No PEA', 'PEA'])
plt.show()

# Crear variable PET (15 a 65 años)
respondieron['PET'] = respondieron['CH06'].apply(lambda x: 1 if 15 <= x <= 65 else 0)

# Gráfico de PET vs PEA
pet_pea = respondieron.groupby(['AÑO', 'PET', 'PEA']).size().unstack().fillna(0)
pet_pea.groupby(level=0).sum().plot(kind='bar', stacked=True)
plt.title("Comparación PEA dentro de PET (15-65 años) por año")
plt.xlabel("Año")
plt.ylabel("Cantidad de personas")
plt.legend(['No PEA', 'PEA'])
plt.show()

# Crear variable desocupado
respondieron['desocupado'] = respondieron['ESTADO'].apply(lambda x: 1 if x == 2 else 0)

# Cantidad de desocupados por año
desocupados_anio = respondieron.groupby('AÑO')['desocupado'].sum()
print("Cantidad de desocupados por año:")
print(desocupados_anio)

# Proporción de desocupados por nivel educativo
nivel_educ_desocup = respondieron.groupby(['AÑO', 'NIVEL_ED'])['desocupado'].mean().unstack()
nivel_educ_desocup.plot(kind='bar')
plt.title("Proporción de desocupados por nivel educativo (2004 vs 2024)")
plt.xlabel("Año")
plt.ylabel("Proporción")
plt.show()

# Crear categorías de edad agrupadas de a 10 años
respondieron['edad_grupo'] = pd.cut(respondieron['CH06'], bins=range(0, 101, 10), right=False)

# Proporción de desocupados por edad agrupada
edad_desocup = respondieron.groupby(['AÑO', 'edad_grupo'])['desocupado'].mean().unstack()
edad_desocup.T.plot(kind='bar')
plt.title("Proporción de desocupados por edad (2004 vs 2024)")
plt.xlabel("Grupo de edad")
plt.ylabel("Proporción de desocupados")
plt.show()

# Análisis libre con otra variable de interés (por ejemplo: CH04 - Sexo)
sexo_desocup = respondieron.groupby(['AÑO', 'CH04'])['desocupado'].mean().unstack()
sexo_desocup.plot(kind='bar')
plt.title("Proporción de desocupados por sexo (2004 vs 2024)")
plt.xlabel("Año")
plt.ylabel("Proporción de desocupados")
plt.legend(['Varón', 'Mujer'])
plt.show()
```
